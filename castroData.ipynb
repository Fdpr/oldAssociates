```python
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
import numpy as np

castro_norms = pd.read_csv('./data/castro_cat_norms.csv')
```

```python
castro_norms.head()
```

```python
castro_norms.YA_Rank.astype(float) - castro_norms.OA_Rank.astype(float)
```

```python
castro_norms = castro_norms.assign(
    differenceRankYAOA = castro_norms.YA_Rank.fillna(0) - castro_norms.OA_Rank.fillna(0),
    differenceRankYAMA = castro_norms.YA_Rank.fillna(0) - castro_norms.MA_Rank.fillna(0),
)
```

```python
castro_norms.sort_values("differenceRankYAOA").head()
```

```python
castro_norms.sort_values("differenceRankYAOA", ascending=False).head()
```

```python
castro_norms.dropna().sort_values("differenceRankYAMA", ascending=True).head()
```

```python
castro_norms.dropna().sort_values("differenceRankYAMA", ascending=False).head()
```

```python
castro_norms.groupby("Cue").sort_values("differenceRankYAOA")
```
